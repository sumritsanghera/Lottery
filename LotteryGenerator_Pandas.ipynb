```python
import pandas as pd
import numpy as np
```

```python
#lottery data from https://www.getmylottoresults.com/lotto-649-past-winning-numbers/
#lotto data from 2021

df = pd.read_csv('lotto649_2021.csv', delimiter = ',')
```

## Code

```python
i = 0
#repeat 500 times, for good measure, and some luck
while i < 500:
    rand1 = np.random.choice(df['1'], 1)
    rand2 = np.random.choice(df['2'], 1)
    rand3 = np.random.choice(df['3'], 1)
    rand4 = np.random.choice(df['4'], 1)
    rand5 = np.random.choice(df['5'], 1)
    rand6 = np.random.choice(df['6'], 1)
    randBonus = np.random.choice(df['Bonus'], 1)
    i+=1

lotteryList = [rand1,rand2,rand3,rand4,rand5,rand6]
lotteryList.sort()

#in case 2 numbers are the same, which is not allowed

for i in range(len(lotteryList)):
    if lotteryList[i-1] == lotteryList[i]:
        if lotteryList[i] == 1:
            lotteryList[i] += 1
        elif lotteryList[i] == 49:
            lotteryList[i] -= 1
        else:
            lotteryList[i] += 1
    
```

# Lottery Numbers:

```python
print("Your numbers are: ")
for j in range(len(lotteryList)):
    print(lotteryList[j], end="")
print("\nGood Luck!")
```
